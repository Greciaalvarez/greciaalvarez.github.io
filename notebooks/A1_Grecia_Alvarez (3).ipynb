# **Introduction**
The purpose of this report is to apply course concepts to develop and evaluate two promising options trading strategies using real option chain data. The analysis is conducted in Python and includes data cleaning, strategy formulation, risk/reward evaluation, profitability metrics and optimization.

In order to evaluate the best strategy, the overall direction of the market was first analyzed. This report is focus on two bullish strategies:

**Bull Call Spread**, which profits from moderate upward price movements while limiting both gains and losses.

**Covered Call**, a conservative income-generating strategy suited for stocks already held.

Both strategies are evaluated based on profitability, risk profile, sensitivity to market parameters (including implied volatility), as well as trade feasibility, liquidity, bid-ask spreads, trading volume, and open interest.

## **Data Cleaning and Preparation**
The dataset provided includes option chain data for a specific underlying stock, with contracts expiring on January 19, 2024. It contains key information such as strike prices, bid/ask prices, implied volatility, and the Greeks (Delta, Gamma, Theta, Vega).

An initial review revealed several missing or inconsistent values, particularly in low-liquidity contracts. To ensure the accuracy and reliability of the analysis, the dataset was cleaned and filtered based on the following criteria:

1. **Missing Values:** Contracts lacking bid, ask, implied volatility, or Greek values were removed.

2. **Bid and Ask Prices:** Only contracts with bid and ask prices above $0.01 were retained to filter out illiquid options.

3. **Open Interest:** A threshold of open interest greater than 5 was applied to focus on contracts with adequate market activity.

4. **At-the-Money (ATM) Selection:** Options with Delta values close to 0.5 were selected to capture ATM contracts, which are typically more responsive to price movements.

5. **Expiration Range:** Contracts expiring within 7 to 14 days were included to align with the short-term focus of the strategies.

The resulting dataset centers on liquid call options near the current stock price and within a defined expiration window, providing a solid foundation for strategy evaluation.

## **Implied Volatility Smile Analysis**

The scatterplot of Implied Volatility (IV) versus Strike Price, colored by Expiration Date, reveals a clear volatility smile pattern. **According to Hull (2018)**, the smile volatility reflects that implied volatility is not constant and depends on the strike and expiration due to market demand and risk perception.

From the chart:
*  **Lower strike prices,** show significantly higher implied volatility, in some cases exceeding 3.5, suggesting the market is pricing in hight uncertainty or demand for protection at those levels.
*  **Higher strike prices,** tend to show lower implied volatility, often below 1.0, indicating a lower expectation of extreme upward movement or less demand for those strikes.

## **Strategy Formulation**
This section will explain the two selected options trading strategies after cleaning and filtering the data.

#### **Bull Call Spread Strategy:**
This strategy is known as a long call spreads, which consist of two positions: buying a call option and selling a call option at a higher strike price, both with the same expiration date.

Was chosen due to the current moderately bullish market utlook. It limits both potential gains and losses, making it suitable for controlled-risk scenarios.

*   **Structure:** Buy 1 ATM Call at a 560.00 strike, and sell 1 OTM Call at a 600 strike,both expiring on 2024-01-19. The spread between the strike prices defines the maximum gain.

*   **Contract Selection:** Considering the current underlying price of $565, the first trade position was slightly at-the-money (ATM), due to its strong volume, open interest, and favorable implied volatility of 0.67. While the second position is out of the money (OTM), chosen for its high open interest and liquidity, and implied volatility of 0.68.

*   **Expected Behavior:** Maximum profit is achieved if the underlying asset closed at or above the short strike (600). If implied volatility (IV) increases, generally raises the value of the spread, benefiting the position. In constrast, a decrease in implied volatility may reduce the value of the spread. In this specific trade, where a bull market is expected, the implied volatility was 0.67 for the long call and 0.68 for the short call, reflect strong market expectations of future movement. The hight (IV) benefits the short position, since It could be sold in a better price, and adds potential value to the long call.

#### **Covered Call:**
The Covered Call is a conservative strategy involving holding the underlying stock and selling a call option to generate additional income. This strategy has a undefined risk and limited profit potential,

*   **Structure:** Hold 100 shares of the underlying stock (Currently priced at $ 565) and sell 1 call option (OTM) at 620 strike against it, expering on 2024 - 01- 19.

*   **Contract Selection:** The short call was selected out of the money to allow balance between limited upside potential and premium income. The contract shows strong market activity with a volume of 189, open interest of 382, and an implied volatility of 0.69. The high IV indicate a strong market expectations.

*   **Expected Behavior:** The maximun profit will be gain is the stock rises above 620 but is limited. However if the stock remains below $620 at expiration, the option expires worthless and the full premium is retained as income. If the stock price drops, the investor absorbs the full loss from the underlying, partially offset by the premium collected.







## **Comparative Analysis and Optimization**
**Insight One: Probability of Profit and Profitability Profile**

The Covered Call strategy has a higher probability of making money (61.97%) than the Bull Call Spread (41.91%). This is because the Covered Call receives income from the start from the premium paid, which reduces the break-even point and makes money even if the price of the underlying asset does not increase. In contrast, the Bull Call Spread requires the price of the underlying asset to exceed a threshold (Breakeven 576.71) to enter the profit zone. Although the Bull Call has a lower initial cost (-$1,670 vs. -$53,770), its returns are more limited and depend more on upward movement. Therefore, in terms of probability and stability, the Covered Call offers a more conservative and predictable profile.

**Insight Two: Sensitivity to Implied Volatility and Greeks**

Bull Call Spread has higher exposure to volatility because both positions (buying and selling calls) have positive Vega: 0.44 and 0.42 respectively. Thats means that if implied volatility goes up, the net value of the strategy will also go up, which is desirable in an uncertain market. Conversely, in the Covered Call, just the sold call is sensitive (Vega of 0.37), whereas the underlying action doesn't move with volatility changes. In addition, Gamma and Theta values reflect that the Bull Call Spread is more exposed to rapid price change and time passing (net Theta -0.09), whereas the Covered Call has a positive Theta, it gains value as time passes, perfect in sideway markets.

**Insight Three: Risk and Expected Return**

While the Bull Call Spread has a predetermined risk (maximum loss = -1,670), the maximum gain is also capped at 2,330, creating a favorable risk-reward ratio for moderately bullish scenarios. However, the Covered Call Spread has no risk limit, as the stock price can fall infinitely, generating larger potential losses (up to -25,520). Moreover, its maximum gain is higher (8,230), provided the price remains below the strike price. This strategy seeks to generate additional income in stable or moderately bullish markets, but it is not suitable if high negative volatility is expected.

## **Risk Management**
**Bull Call Spread Strategy**

*   IV Increase:

An increase in implied volatility has the effect of increasing the value of both the long and the short calls. However, since the long call is closer to at-the-money (ATM), it is more sensitive to IV changes. As a result, the net value of the spread increases, and the profit potential can become higher. This is favorable to the position.

*   IV Decrease:

A drop in implied volatility reduces the premiums of both options. The long call (buy) loses more value than the short call (sell), making the net spread value go down. This reduces the profitability, especially if the underlying price is moving slowly.

*   Volatility Sensitivity:

The Bull Call Spread is Vega-positive, though moderately. It benefits from rising volatility, but presents less exposure than a direct call, making it a risk-controlled strategy.

**Covered Call Strategy**

*   IV Increase:

An increase in implied volatility raises the premium of the short call. But it is generally a drawback, as it increases the probability that the stock price will rise above the strike price, capping gains. And wider volatility can lead to broader stock price oscillations, raising downside risk. The position is more volatile overall.

*   IV Decrease:

A drop in implied volatility reduces the premium received but also minimizes the risk of early assignment and price movement. Since the goal is to keep the option premium as revenue, a fall in IV is favorable to this strategy.

*   Volatility Sensitivity:

The Covered Call is Vega-negative, it benefits from stable or falling volatility, in line with its conservative approach.

## **Conclusion**
Based on a comparison of the two bullish options strategies, **Bull Call Spread** and **Covered Call**, this research presents an analysis of how they behave under the current market conditions and compares their advantages in profitability, risk, and volatility sensitivity.

The Bull Call Spread provides a risk-managed opportunity to gain from modest price advances and with pre-defined maximum loss it give a favorable risk-reward structure make it an appropriate choice in moderately bullish conditions, especially where implied volatility stands to rise. Although less likely to be profitable than the Covered Call, this trade generates capital efficiency and defined downside risk.

In contrast, the Covered Call has a more conservative profile. It returns consistent premium income and has a much greater chance of making a profit, for which it is especially well suited in flat or weakly bullish markets. The investor is, however, also vulnerable to potentially unlimited losses on the downside if the underlying stock drops sharply. It can take advantage of time decay and falling or stable volatility, typical of a low-risk, income-accreting strategy.

From the risk management point of view, the Bull Call Spread is quite Vega-positive and more market-sensitive, while the Covered Call is Vega-negative and best used in low volatilities.  

**Suggestion to the Firm:**

Both strategies serve distinct purposes and cater to different investor risk profiles. For more risk-tolerant investors or clients who have a bullish outlook on the market, the **Bull Call Spread** is the favored strategy due to its defined risk and potential for moderate gains. On the other hand, for more risk-averse clients who seeking for stable returns and hold underlying assets, the **Covered Call** is a good and reliable strategy. A blended strategy may also be applied to diversify the risk and react to various market conditions.

```python
import datetime as dt
import pandas as pd
from optionlab import run_strategy, plot_pl
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# for interactive tables
from google.colab import data_table
data_table.enable_dataframe_formatter()
```

```python
# example option chain data
fcsv = "https://raw.githubusercontent.com/multidis/hult-options-trading/refs/heads/main/option_chain_data/2023-12-11_option_chain_120days_student13.csv"
dfchain = pd.read_csv(fcsv)
```

```python
dfchain
```

```python
# some contracts may have NA-values
dfchain[dfchain.isna().any(axis=1)]
```

```python
filt_na = ~dfchain.isna().any(axis=1)
filt_prc = (dfchain['bid'] > 0.01) & (dfchain['ask'] > 0.01)
filt_oi = dfchain['open_interest'] > 5
```

```python
dffilt = dfchain[filt_na & filt_prc & filt_oi]
```

```python
dffilt
```

```python
# underlying price near ATM contracts
atm_contracts = dffilt[abs(dffilt['delta'] - 0.5) < 0.015]
```

```python
atm_contracts
```

```python
und_price = 565
start_date = dt.date(2024, 1, 5 )
```

```python

dffilt.loc[:, 'expiration_date'] = pd.to_datetime(dffilt['expiration_date']).dt.date
```

```python
dffilt_exp = dffilt[
    (dffilt['expiration_date'] > start_date + dt.timedelta(days=7)) &
    (dffilt['expiration_date'] <= start_date + dt.timedelta(days=45))
]
```

```python
dffilt_exp
```

```python
# Simulating Different Market Scenarios
inter = 0.05
und_min = und_price - round(0.5*und_price, 2)
und_max = und_price + round(0.5*und_price, 2)
```

```python
# contract indices to use in the strategy
isell = 1634 #Index
ibuy =  1626  #Index
end_date = dt.date(2024, 1, 19)
```

```python
# strategy for Bullish - Bull Call Spread

strategy = [
    {
        "type": "call",
        "strike": dffilt_exp.loc[ibuy, 'strike'],
        "premium": dffilt_exp.loc[ibuy, 'ask'],
        "n": 100, # number of contracts
        "action": "buy",
        "expiration": dffilt_exp.loc[ibuy, 'expiration_date']
    },
    {
        "type": "call",
        "strike": dffilt_exp.loc[isell, 'strike'],
        "premium": dffilt_exp.loc[isell, 'bid'],
        "n": 100, # number of contracts
        "action": "sell",
        "expiration": dffilt_exp.loc[isell, 'expiration_date'],
    }
]
strategy
```

```python
inputs = {
    "stock_price": und_price,
    "start_date": start_date,
    "target_date": end_date,
    "volatility": dfchain.loc[isell, 'mid_iv'],
    "interest_rate": inter,
    "min_stock": und_min,
    "max_stock": und_max,
    "strategy": strategy
}
inputs
```

```python
st = run_strategy(inputs)
```

```python
plt.title('Profit/Loss Diagram - Bull Call Spread')
plot_pl(st)
```

```python
print(st)
```

```python
st.probability_of_profit
```

```python
st.minimum_return_in_the_domain
```

```python
st.maximum_return_in_the_domain
```
